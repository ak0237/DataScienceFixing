```python
import pandas as pd

import matplotlib.pyplot as plt
from matplotlib import rcParams
import seaborn as sns

%matplotlib inline
rcParams['figure.figsize'] = 8, 4
sns.set_style('whitegrid')
```

## Supondo a distribuição dos dados com histogramas

```python
caminho = 'C:/Users/Usuario/Documents/arquivos_de_exercicios_python_para_ciencia_dados_basico/dados/mtcars.csv'
carros = pd.read_csv(caminho)
carros.columns = ['nomes','mpg','cyl','disp', 'hpt', 'drat', 'wt', 'qsec', 'vs', 'am', 'gear', 'carb']
carros.index = carros.nomes
mpg = carros['mpg']

plt.hist(mpg)
plt.show()
```

**Histograma Seaborn**


```python
sns.histplot(mpg, kde=True, stat="density",  lw=0)
```

## Scatterplots

```python
carros.plot(kind='scatter', x="hpt", y="mpg", c = ['blue'], s = 100 )
```

**SeaBorn**

```python
sns.regplot(x='hpt', y='mpg', data=carros, scatter=True)
```

## Gerando uma matriz de scatter plots

```python
sns.pairplot(carros)
```

```python
colunas = ['mpg', 'disp', 'hpt','wt']
novo_df = pd.DataFrame((carros[colunas].values), columns=colunas)
carros_target = carros['am'].values

novo_df['grupo'] = pd.Series(carros_target, dtype='category')
```

```python
sns.pairplot(novo_df, hue='grupo', palette='hls')
```

## Construindo boxsplots

```python
carros.boxplot(column='mpg', by='am')
carros.boxplot(column='wt', by='am')
```

**SeaBorn**

```python
sns.boxplot(x='am', y = 'mpg', data=carros, palette='hls')
```
