```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
from pylab import rcParams

import seaborn as sns


%matplotlib inline
rcParams['figure.figsize'] = 5, 4
sns.set_style('whitegrid')
```

## Definindo uma cor

```python
x = range(1,10)
y = [1, 2, 0, 2, 2, 1, 4, 5, 9]

plt.bar(x, y)
```

```python
cor = ['salmon']
plt.bar(x, y, color = cor)
```

## Customizando estilos das linhas

```python
x1 = range(0, 10)
y1 = [1 ,5, 9, 7, 5, 3, 2, 4, 6, 8]

plt.plot(x,y)
plt.plot(x1, y1)
```

```python


# ds = drawstyle  lw = tamanho  'steps' = "escadinha"  '--' = tracejado

plt.plot(x, y, ds='steps', linewidth=5)
plt.plot(x1, y1, '--', lw= 3)
```

## Definido marcadores em gr√°ficos

```python
plt.plot(x, y, marker='1', mew=20)
plt.plot(x1, y1, marker ='+', new=15)
```
