```python
import pandas as pd
import matplotlib.pyplot as ptl
from matplotlib import rcParams

import seaborn as sns

%matplotlib inline
rcParams['figure.figsize'] = 6, 3
sns.set_style('whitegrid')

caminho = 'C:/Users/Usuario/Documents/arquivos_de_exercicios_python_para_ciencia_dados_basico/dados/mtcars.csv'
carros = pd.read_csv(caminho)
carros.columns = ['nomes','mpg','cyl','disp', 'hpt', 'drat', 'wt', 'qsec', 'vs', 'am', 'gear', 'carb']
```

```python
from scipy.stats.stats import pearsonr
```

## Correlação de Pearson

```python
x = carros[['mpg', 'hpt', 'wt', 'qsec']]
sns.pairplot(x)
```

```python
mpg = carros['mpg']
hpt = carros['hpt']
qsec = carros['qsec']
wt = carros['wt']

coeficiente_pearson, p_valor = pearsonr(mpg, hpt)
print('Coeficiente de Pearson {:0.3}'.format(coeficiente_pearson))
```

```python
coeficiente_pearson, p_valor = pearsonr(mpg, wt)
print('Coeficiente de Pearson {:0.3}'.format(coeficiente_pearson))
```

```python
coeficiente_pearson, p_valor = pearsonr(mpg, qsec)
print('Coeficiente de Pearson {:0.3}'.format(coeficiente_pearson))
```
