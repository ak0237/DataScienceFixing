```python
import pandas as pd

import matplotlib.pyplot as plt
from matplotlib import rcParams
import seaborn as sns

%matplotlib inline
rcParams['figure.figsize'] = 20, 4
sns.set_style('whitegrid')

caminho = 'C:/Users/Usuario/Documents/arquivos_de_exercicios_python_para_ciencia_dados_basico/dados/mtcars.csv'

carros = pd.read_csv(caminho)
carros.columns = ['nomes','mpg','cyl','disp', 'hpt', 'drat', 'wt', 'qsec', 'vs', 'am', 'gear', 'carb']
mpg = carros['mpg']
```

## Colocando t√≠tulo nos eixos

```python
x = range(0, 10)
y = [1,2,3,4,6,7,9,5,8,10]
plt.bar(x,y)
plt.xlabel('Eixo x')
plt.ylabel('Eixo y')
```

## Legendas

```python
import matplotlib.patches as mpatches

plt.plot(mpg)
plt.xlabel('Nomes dos Carros')
plt.ylabel('Milhas/Gal√£o')
plt.title('Consumo de Gasolina')

plt.xticks(mpg.index, carros.nomes, rotation=80)

legenda = mpatches.Patch(label='mpg')
plt.legend(handles=[legenda])
```
